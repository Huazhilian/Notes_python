**plot curves in one section**

```python
import numpy as np
from matplotlib import pyplot as plt
from math import cos,sin,pi

x=np.linspace(-pi,pi,100)
y=list(map(sin,x))
z=list(map(cos,x))

plt.title('cos(x) & sin(x)')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.plot(x,y,'g',label='sin(x)')
plt.plot(x,z,'y',label='cos(x)')
plt.legend()
plt.show()
```

**plot curves in subsections**


```python
import numpy as np
from matplotlib import pyplot as plt
from math import exp

x=np.linspace(-1,1,100)
y=list(map(exp,x))
z=x**2+1

plt.subplot(2,1,1)
plt.plot(x,y,'y',label='exp(x)')
plt.title('y=exp(x)')
plt.legend()

plt.subplot(2,1,2)
plt.plot(x,z,'c',label='x^2+1')
plt.title('y=x^2+1')
plt.legend()
plt.show()
```

- ### Bar chart

**plot bar chart**


```python
import numpy as np
from matplotlib import pyplot as plt

x1=np.arange(1,13,2)
y1=np.array([15,9,2,4,6,8])
x2=np.arange(2,13,2)
y2=np.array([5,3,9,12,7,6])

plt.bar(x1,y1,align='center')
plt.bar(x2,y2,align='center')
plt.title('Bar chart')
plt.xlabel('Month')
plt.ylabel('Volume')
plt.show()
```

- ### Histogram

**plot histogram**


```python
import numpy as np
from matplotlib import pyplot as plt

data=np.array([2,3,7,3,8,9,5,0,4,7,1,5,6,6,6])
interval=np.arange(0,10)
plt.hist(data,bins=interval)
plt.title('histogram')
plt.show()
```
