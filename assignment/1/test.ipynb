```python
import csv
import numpy as np
from matplotlib import pyplot as plt

with open(r'C:\Users\Taylor\OneDrive\Desktop\Code\Python\Notes\assignment\Data_1min.csv', 'r') as file:
    # data handling
    reader = csv.reader(file)
    data = np.array(list(reader))
    category = data[0] 
    data = data[1:].T
    data = [data[i][:100] for i in range(len(data))]  # Limit to first 100 entries
    length = len(data[0])

    # data visualization
    print(category)
    print(len(data))
    x = np.arange(length)
    for i in range(1,len(data)-2):
        plt.plot(x, data[i],label=category[i])
        plt.xlabel('Time (1 minute intervals)')
        plt.ylabel('price')
        plt.legend()
        plt.title('Price Data Over Time')
        plt.show()

    # moving average calculation
    moving_average = np.convolve(data[1].astype(float), np.ones(10)/10, mode='valid')
    plt.plot(x[9:], moving_average, label='Moving Average', color='orange')
    plt.xlabel('Time (1 minute intervals)')
    plt.ylabel('Price')
    plt.legend()
    plt.title('Moving Average of Price Data')
    plt.show()
```
